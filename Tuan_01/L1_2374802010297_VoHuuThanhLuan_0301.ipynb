# Tải dữ liệu xét tuyển đại học

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col="STT")
df.tail(10)
```

```python
df.head(10)
```

```python
# Thông kê dữ liệu thiếu  cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu thì điền giá trị sử dụng phương pháp Mean.

df['DT'].fillna(0, inplace=True)
df
```

```python
df["T1"].fillna(df["T1"].mean(), inplace = True)
df
```

```python
#6. Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.
score_cols = df.select_dtypes(include=[np.number]).columns
df[score_cols] = df[score_cols].fillna(0)
print(df['DT'].head(10))
```

```python
#7. Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12.
for i in [1, 2, 3]:
    df[f'TBM{i}'] = (df[f'T{i}']*2 + df[f'L{i}'] + df[f'H{i}'] + df[f'S{i}'] +
                     df[f'V{i}']*2 + df[f'X{i}'] + df[f'D{i}'] + df[f'N{i}']) / 10

print(df[['TBM1', 'TBM2', 'TBM3']].head())
```

```python
#8. Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp
#10, 11, 12 như sau:
#• Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
#• Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
#• Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
#• Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
#• Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)

def xep_loai(diem):
    if diem < 5.0:
        return 'Y'
    elif 5.0 <= diem < 6.5:
        return 'TB'
    elif 6.5 <= diem < 8.0:
        return 'K'
    elif 8.0 <= diem < 9.0:
        return 'G'
    else:
        return 'XS'

for i in [1, 2, 3]:
    df[f'XL{i}'] = df[f'TBM{i}'].apply(xep_loai)

print(df[['TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head())
```

```python
#9. Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp
#10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương
#pháp Min-Max Normalization

for i in [1, 2, 3]:
    col_name = f'TBM{i}'
    min_val = df[col_name].min()
    max_val = df[col_name].max()


    if max_val - min_val != 0:
        df[f'US_TBM{i}'] = (df[col_name] - min_val) / (max_val - min_val) * 4
    else:
        df[f'US_TBM{i}'] = 0.0

print(df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head())
```

```python
#10. Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và
#rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3 như sau
#• Với khối A, A1 nếu [(DH1*2 + DH2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu,
#ngược lại là rớt
#• Với khối B nếu [(DH1 + DH2*2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược
#lại là rớt
#• Với khối khác nếu [(DH1+ DH2 + DH3)/3] lớn hơn hoặc bằng 5.0 thì đậu, ngược
#lại là rớt

def xet_tuyen(row):
    khoi = row['KT']
    dh1 = row['DH1']
    dh2 = row['DH2']
    dh3 = row['DH3']

    dtb = 0
    if khoi in ['A', 'A1']:
        dtb = (dh1*2 + dh2 + dh3) / 4
    elif khoi == 'B':
        dtb = (dh1 + dh2*2 + dh3) / 4
    else:
        dtb = (dh1 + dh2 + dh3) / 3

    return 1 if dtb >= 5.0 else 0

df['KQXT'] = df.apply(xet_tuyen, axis=1)

print(df[['KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head(10))
```

```python
#11. Lưu trữ dữ liệu xuống ổ đĩa thành file processed_dulieuxettuyendaihoc.csv
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
```
