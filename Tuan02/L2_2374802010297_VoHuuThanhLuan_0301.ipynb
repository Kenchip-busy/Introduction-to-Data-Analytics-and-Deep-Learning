```python
#1. Vẽ biểu đồ đường Simple cho biến T1
dftuyensinh['T1'].plot(title='Biểu đồ đường (Line Plot) cho biến T1', figsize=(10, 6))
plt.ylabel('Điểm T1')
plt.xlabel('Số thứ tự sinh viên')
plt.show()
```

```python
#2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:
def phan_loai(score):
    if score < 5: return 'kem'
    elif score < 7: return 'tb'
    elif score < 8: return 'kha'
    else: return 'gioi'

dftuyensinh['phanlopt1'] = dftuyensinh['T1'].apply(phan_loai)
```

```python
#3. Lập bảng tần số cho biến phanloait1
print("Bảng tần số phân loại T1:")
print(dftuyensinh['phanlopt1'].value_counts())
```

```python
#4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1
# Thiết lập kích thước biểu đồ to rõ
grouped = dftuyensinh.groupby(['T1', 'phanlopt1']).size().unstack(fill_value=0)

# 2. Vẽ biểu đồ (Đã bỏ marker)
plt.figure(figsize=(10, 6))
for col in grouped.columns:
    plt.plot(grouped.index, grouped[col], label=col)

plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Multiple Line cho biến T1 được phân loại bởi biến phanlopt1")
plt.legend(title="PhanlopT1")
plt.grid(True)
plt.show()
```

```python
#5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1
fig, ax = plt.subplots(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'orange']
for i, col in enumerate(grouped.columns):
    markerline, stemlines, baseline = ax.stem(grouped.index, grouped[col], label=col)
    color = colors[i % len(colors)]
    plt.setp(stemlines, 'color', color)
    plt.setp(markerline, 'markerfacecolor', color, 'markeredgecolor', color, 'marker', 'o') # 'o' là hình tròn

plt.legend(title='Phân lớp T1')
plt.xlabel('Điểm T1')
plt.ylabel('Số lượng (Count)')
plt.title('Biểu đồ Drop-line cho biến T1 (Chấm tròn)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#1. Hãy mô tả và khảo sát phân phối cho biến T1
print("\n=== CÂU 1: PHÂN TÍCH BIẾN T1 ===")
# 1.1 Mô tả độ tập trung và phân tán
desc = dftuyensinh['T1'].describe()
skewness = dftuyensinh['T1'].skew()
kurtosis = dftuyensinh['T1'].kurtosis()

print("Thống kê mô tả:")
print(desc)
print(f"Skewness (Độ lệch): {skewness:.4f}")
print(f"Kurtosis (Độ nhọn): {kurtosis:.4f}")

# 1.2 Vẽ biểu đồ (Box-plot, Histogram, QQ-Plot)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Box-Plot
sns.boxplot(x=dftuyensinh['T1'], ax=axes[0], color='lightblue')
axes[0].set_title('Box-Plot biến T1')

# Histogram
sns.histplot(dftuyensinh['T1'], kde=True, ax=axes[1], color='green')
axes[1].set_title('Histogram biến T1')

# QQ-Plot (Kiểm tra phân phối chuẩn)
stats.probplot(dftuyensinh['T1'], dist="norm", plot=axes[2])
axes[2].set_title('QQ-Plot biến T1')

plt.tight_layout()
plt.show()

# Nhận xét phân phối
print("\n--- NHẬN XÉT CÂU 1 ---")
if abs(skewness) < 0.5:
    print(f"- Skewness = {skewness:.2f}: Phân phối khá đối xứng (gần chuẩn).")
elif skewness > 0:
    print(f"- Skewness = {skewness:.2f}: Phân phối lệch phải (Positive Skew).")
else:
    print(f"- Skewness = {skewness:.2f}: Phân phối lệch trái (Negative Skew).")
```

```python
#2. Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)
plt.figure(figsize=(10, 6))
sns.boxplot(x='phanlopt1', y='T1', data=dftuyensinh,
            order=['kem', 'tb', 'kha', 'gioi'], palette='Set3')
plt.title('Box-Plot T1 phân theo nhóm xếp loại')
plt.show()

# 2.2 Histogram theo phân lớp
g = sns.FacetGrid(dftuyensinh, col="phanlopt1", col_order=['kem', 'tb', 'kha', 'gioi'], col_wrap=2)
g.map(sns.histplot, "T1", kde=True)
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Histogram T1 theo từng nhóm')
plt.show()

# 2.3 QQ-Plot theo phân lớp
def qqplot_group(x, **kwargs):
    ax = plt.gca()
    stats.probplot(x, dist="norm", plot=ax)

g = sns.FacetGrid(dftuyensinh, col="phanlopt1", col_order=['kem', 'tb', 'kha', 'gioi'], col_wrap=2)
g.map(qqplot_group, "T1")
plt.subplots_adjust(top=0.9)
g.fig.suptitle('QQ-Plot T1 theo từng nhóm')
plt.show()
```

```python
#3. Hãy khảo sát tương quan giữa biến DH1 theo biến T1
corr_dh1_t1 = dftuyensinh['DH1'].corr(dftuyensinh['T1']) # Hệ số tương quan
cov_dh1_t1 = dftuyensinh['DH1'].cov(dftuyensinh['T1'])   # Hiệp phương sai

print(f"Hệ số tương quan (Correlation): {corr_dh1_t1:.4f}")
print(f"Hiệp phương sai (Covariance): {cov_dh1_t1:.4f}")

# 3.2 Vẽ Scatter Plot kèm đường hồi quy tuyến tính
plt.figure(figsize=(8, 6))
sns.regplot(x='T1', y='DH1', data=dftuyensinh, line_kws={'color': 'red'})
plt.title(f'Biểu đồ Scatter: T1 vs DH1 (r={corr_dh1_t1:.2f})')
plt.xlabel('Điểm T1 (Học bạ)')
plt.ylabel('Điểm DH1 (Điểm thi)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#4. Hãy khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực
sns.lmplot(x='T1', y='DH1', hue='KV', data=dftuyensinh, height=6, aspect=1.5)
plt.title('Tương quan T1 và DH1 phân theo Khu vực (KV)')
plt.show()
```

```python
#5. Hãy khảo sát tương quan giữa các biến DH1, DH2, DH3
cols = ['DH1', 'DH2', 'DH3']

# 5.1 Ma trận tương quan
print("Ma trận tương quan (Correlation Matrix):")
print(dftuyensinh[cols].corr())

# 5.2 Ma trận hiệp phương sai
print("\nMa trận hiệp phương sai (Covariance Matrix):")
print(dftuyensinh[cols].cov())

# 5.3 Biểu đồ Pairplot (Vẽ tương quan chéo giữa tất cả các cặp biến)
sns.pairplot(dftuyensinh[cols])
plt.suptitle('Biểu đồ Pairplot: Quan hệ giữa điểm thi 3 môn', y=1.02)
plt.show()
```
